```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle

RANDOM_SEED = 42
```

```python
DATADIR = "dataset/training"
TRAIN_DATADIR = "dataset/training"
TEST_DATADIR = "dataset/test"
LETTERS = os.listdir(DATADIR)
```

```python
data = []
X = []
y = []

for letter in LETTERS:
    for image in os.listdir((TRAIN_DATADIR + '/' + letter)):
        IMAGE_DIR = (TRAIN_DATADIR + '/' + letter + '/' + image)
        image_array = cv2.imread(IMAGE_DIR)
        data.append([image_array, LETTERS.index(letter)])
        print(IMAGE_DIR)
        
    for image in os.listdir((TEST_DATADIR + '/' + letter)):
        IMAGE_DIR = (TEST_DATADIR + '/' + letter + '/' + image)
        image_array = cv2.imread(IMAGE_DIR)
        data.append([image_array, LETTERS.index(letter)])
        print(IMAGE_DIR)
        
# Corrigindo BUG da leitura (no tava convertendo BGR2RGB na leitura da imagem)      
print_ = True
for tupla in data:
    original_img = cv2.cvtColor(tupla[0], cv2.COLOR_BGR2RGB)
    brightness_img = cv2.convertScaleAbs(original_img, beta=60) #ajusta brilho
    
    kernel = np.array([[-1, -1, -1], [-1, 9.5, -1], [-1, -1, -1]])
    sharpen_img = cv2.filter2D(original_img, -1, kernel)
    
    X.append(original_img)
    X.append(brightness_img)
    X.append(sharpen_img)
    
    if print_ == True:
        plt.imshow(original_img)
        plt.show()
        
        plt.imshow(brightness_img)
        plt.show()
        
        plt.imshow(sharpen_img)
        print_ = False
    
    
    y.append(tupla[1])
    y.append(tupla[1])
    y.append(tupla[1])
    
X = np.asarray(X)
y = np.asarray(y)
```

```python
plt.imshow(data[0][0])
plt.title(data[0][1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split


# O X_train e X_test j retornam em formato de tensor!! AMO VC SKLEARN <3 <3 <3 <3 <3 <3 <3 <3 <3 <3 <3 <3 <3 <3 <3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```
